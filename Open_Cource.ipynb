```python
for i in nodelist:
    for j in i:
        print(j,end = " ")
    print()

jumlahLokasi =  len(nodelist)   
print(jumlahLokasi)
```

###Inisialisasi Populasi secara acak

###Array P diinisialisasi sebagai array integer

###Solusi awal dibuat dengan random permutation


```python
def initPopulasi(ukuranPopulasi,jumlahLokasi):
  P = np.empty((ukuranPopulasi,jumlahLokasi))
  P = P.astype('int')
  for i in range(ukuranPopulasi):
    I = np.random.permutation(jumlahLokasi)
    for j in range(jumlahLokasi):
      P[i][j]=I[j]
  return P

ukuranPopulasi = 10
P = initPopulasi (ukuranPopulasi,jumlahLokasi)
print("Inisialisasi Populasi : ",P)
```

###Membuat Fungsi Objektif

###Jarak yang diukur adalah Jarak Euclidian

```python
def fungsiObjectif(populasi,nodelist):
  sz = populasi.shape
  ukuranPopulasi = sz[0]
  jumlahLokasi = sz[1]

  matrikJarak = np.empty((ukuranPopulasi))
  matrikJarak = matrikJarak.astype('float')
  n = len(nodelist)

  for i in range(ukuranPopulasi):
    jarak = 0
    for j in range(n-1):
      jrkX = nodelist[populasi[i][j]][0]-nodelist[populasi[i][j+1]][0]
      jrkY = nodelist[populasi[i][j]][1]-nodelist[populasi[i][j+1]][1]
      d = np.sqrt(jrkX**2 + jrkY**2 )   
      jarak = jarak  + d
      
    dn =  np.sqrt((nodelist[populasi[i][n-1]][0]-nodelist[populasi[i][0]][0])**2+ (nodelist[populasi[i][n-1]][1]-nodelist[populasi[i][0]][1])**2 )
    jarak = jarak  + dn
    matrikJarak[i] = jarak

  return matrikJarak

fitP = fungsiObjectif(P,nodelist)
print(fitP)
```

###Seleksi menggunakan Metode Tournamen

###Ukuran Tournamen k = 2

###Jumlah Induk yang dipilih adalah 2

```python
def seleksi (P,fsObj):
  jmlInduk = 2
  sz = P.shape
  ukuranPopulasi = sz[0]
  jumlahLokasi = sz[1]

  Induk = np.empty((jmlInduk,jumlahLokasi))
  Induk = Induk.astype('int')
  
  for i in range (2):
    #memilih dua individu secara acak 
    r1 = random.randint(0,ukuranPopulasi-1)
    r2 = random.randint(0,ukuranPopulasi-1)
    
    if  fsObj[r1] < fsObj [r2]:
      for j in range(jumlahLokasi):
        Induk [i][j] = P[r1][j]
    else:
      for j in range(jumlahLokasi):
        Induk [i][j] = P[r2][j]
  return Induk

Induk = seleksi(P,fitP)
print(Induk)
```

###Rekombinasi menggunakan Metode Order Crossover (OX) Jumlah anak yang dihasilkan adalah 2
###Crossover 

```python
def PMX(Induk):
  Induk = seleksi(P,fitP)
  sz = Induk.shape
  brs = sz[0]
  kol = sz[1]

  I1 = np.empty((kol))
  I2 = np.empty((kol))
  A1 = np.empty((kol))
  A2 = np.empty((kol))
  Anak = np.empty((2,kol))
  Anak = Anak.astype('int')

  for i in range(kol):
    I1[i]=Induk[0][i]
    I2[i]=Induk[1][i]

  r1 = random.randint(1,kol-2)

  for i0 in range(r1): 
    A1[i0] = I1[i0]
    A2[i0] = I2[i0]

  batasCek1 = r1
  batasCek2 = r1

  for i in range(kol):
    m = (i+r1)%kol

    n1 = I1[m] 
    n2 = I2[m]

    #mengecek apakah Lokasi tertentu sudah ada di anak

    idxAda1 = 0
    idxAda2 = 0
    for j1 in range(batasCek1): 
      
      if batasCek1>(kol):
        idxAda1 = 1

      if A1[j1] == n2: 
        idxAda1 = 1
            
    for j2 in range(batasCek2): 
      
      if batasCek2>(kol):
        idxAda2 = 1

      if A2[j2] == n1: 
        idxAda2 = 1

    if idxAda1 == 0:
      A1[batasCek1] = n2
      batasCek1 = batasCek1 +1
   
    if idxAda2 == 0:
      A2[batasCek2] = n1
      batasCek2 = batasCek2 +1
   
  for i in range(kol):
    Anak[0][i]=A1[i]
    Anak[1][i]=A2[i]

  return Anak

AnakCX = PMX(Induk)
print("Anak :",AnakCX)
```

###Mutasi dengan Swap

###Index Lokasi yang di Swap pada kedua anak sama

```python
def MutasiSwap(Induk):
  sz = Induk.shape
  brs = sz[0]
  kol = sz[1]

  Anak = np.empty((2,kol))
  Anak = Anak.astype('int')

  Anak = Induk

  r1 = random.randint(0,kol-1)
  r2 = random.randint(0,kol-1)

  while r1 == r2:
    r2 = random.randint(0,kol-1)
  
  if r1 < r2:
    rs1 = r1
    rs2 = r2
  else:
    rs1 = r2
    rs2 = r1

  I1rs1 = Induk [0][rs1]
  I1rs2 = Induk [0][rs2]
  I2rs1 = Induk [1][rs1]
  I2rs2 = Induk [1][rs2]

  Anak[0][rs1] = I1rs2
  Anak[0][rs2] = I1rs1
  Anak[1][rs1] = I2rs2
  Anak[1][rs2] = I2rs1

  return Anak

print("AnakCX :",AnakCX)
AnakM = MutasiSwap(AnakCX)
print("AnakM :",AnakM)
```

###Elitism (steady state replacement)

```python
def Elitism(P,Anak,fitP,fitAnak):
  szP = P.shape
  brsP = szP[0]
  kolP = szP[1]
  szA = Anak.shape
  brsA = szA[0]
  kolA = szA[1]

  #print (fitP)

  for i in range(brsA):
    iJelek = fitP.max()
    idxJelek = fitP.argmax()

    if fitAnak[i]<iJelek:
      fitP[idxJelek] = fitAnak[i]
      for j in range (kolP):
        P[idxJelek][j] = Anak[i][j]
  
  return [P,fitP]

fitAnak = fungsiObjectif(AnakM,nodelist)
print("P :",P)
print("AnakM :",AnakM)
print("fitP :",fitP)
print("fitAnak :",fitAnak)
P = Elitism (P,AnakM,fitP,fitAnak)
print("P :",P)
print("fitP :",fitP)
```

###Prosedur GA

```python
#Inisialisai Parameter
PCX = 0.95
Pm = 0.01
ukuranPopulasi = 10
max_generasi = 5

#Inisialisasi Populasi

P = initPopulasi (ukuranPopulasi,jumlahLokasi)
print("P0 :",P)

#melakukan Evaluasi P

fitP = fungsiObjectif(P,nodelist)
print("fitness populasi awal :",fitP)

print("fitness terbaik awal :",fitP.min())
print("rute awal :",P[fitP.argmin()])

for i in range(max_generasi):
  Induk = seleksi(P,fitP)
  #print("Induk :",Induk)

  Anak = Induk

  rxo = random.random()

  if rxo < PCX:
    Anak = PMX(Induk)

    rm = random.random()
    if rm < Pm:
      Anak = MutasiSwap(Anak)

    fitAnak = fungsiObjectif(Anak,nodelist)
    #print("fitAnak :",fitAnak)

    [P,fitP] = Elitism (P,Anak,fitP,fitAnak)

print("fitness populasi akhir :",fitP)
print("fitness akhir :",fitP.min())
print("rute akhir :",P[fitP.argmin()])
```

```python
def Elitism(P,Anak,fitP,fitAnak):
  szP = P.shape
  brsP = szP[0]
  kolP = szP[1]
  szA = Anak.shape
  brsA = szA[0]
  kolA = szA[1]

  #print (fitP)

  for i in range(brsA):
    iJelek = fitP.max()
    idxJelek = fitP.argmax()

    if fitAnak[i]<iJelek:
      fitP[idxJelek] = fitAnak[i]
      for j in range (kolP):
        P[idxJelek][j] = Anak[i][j]
  
  return [P,fitP]

fitAnak = fungsiObjectif(AnakM,nodelist)
print("P :",P)
print("AnakM :",AnakM)
print("fitP :",fitP)
print("fitAnak :",fitAnak)
P = Elitism (P,AnakM,fitP,fitAnak)
print("P :",P)
print("fitP :",fitP)
```
